## Classic Word Embeddings

```python
from flair.embeddings import WordEmbeddings
print('start: will init / load embeddings now ....')
# init embedding

#glove_embedding = WordEmbeddings('glove')
komninos_embedding = WordEmbeddings('en-extvec') 
#ft_embedding = WordEmbeddings('en') 
#ft_ru_embedding = WordEmbeddings('ru')
print('Done')
```

```python
from flair.data import Sentence

# create sentence.
sentence = Sentence('The grass is green .')

# embed a sentence using glove.
komninos_embedding.embed(sentence)

# now check out the embedded tokens.
for token in sentence:
    print(token)
    #print(token.embedding)
    print('vector length', len(token.embedding), 'dim')
```

## Character-based Embeddings

```python
from flair.embeddings import CharacterEmbeddings

# init embedding
embedding = CharacterEmbeddings()

# create a sentence
sentence = Sentence('The grass is green .')

# embed words in sentence
embedding.embed(sentence)
```

## Byte-pair Embeddings

```python
from flair.embeddings import BytePairEmbeddings

# init embedding
bpe_embedding = BytePairEmbeddings('en')

# create a sentence
sentence = Sentence('The grass is green .')

# embed words in sentence
bpe_embedding.embed(sentence)
```

## Stacked Embeddings

```python
from flair.embeddings import WordEmbeddings, CharacterEmbeddings

# init standard GloVe embedding
glove_embedding = WordEmbeddings('glove')

# init standard character embeddings
character_embeddings = CharacterEmbeddings()

print('init done')
```

```python
from flair.embeddings import StackedEmbeddings

# now create the StackedEmbedding object that combines all embeddings
stacked_embeddings = StackedEmbeddings(
    embeddings=[glove_embedding, character_embeddings, bpe_embedding])
```

```python
sentence = Sentence('The grass is green .')

# just embed a sentence using the StackedEmbedding as you would with any single embedding.
stacked_embeddings.embed(sentence)

# now check out the embedded tokens.
for token in sentence:
    print(token)
    print(token.embedding)
    print('vector length', len(token.embedding), 'dim')
```
